```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
list=os.listdir()
path=os.getcwd()
jpg_files = [f for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
im_1=plt.imread(jpg_files[0])
jpg_files[0]
```

```python
def get_value_from_triple(temp1):
    # temp1=im_1[0,0,:]
    # temp1[0],temp1[1],temp1[2]
    return int(temp1[0]/3+temp1[1]/3+temp1[2]/3)
def get_0_1_from_triple(temp1):
    # temp1=im_1[0,0,:]
    # temp1[0],temp1[1],temp1[2]
    temp=int(temp1[0]/3+temp1[1]/3+temp1[2]/3)
    if temp<140:
        return 0
    else:
        return 1
```

```python
def convert_rgb_to_gray(im_1):
    m,n,k=im_1.shape
    new_image=np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            s=get_value_from_triple(im_1[i,j,:])
            new_image[i,j]=s
    return new_image
def convert_rgb_to_bw(im_1):
    m,n,k=im_1.shape
    new_image=np.zeros((m,n),dtype='uint8')
    for i in range(m):
        for j in range(n):
            s=get_0_1_from_triple(im_1[i,j,:])
            new_image[i,j]=s
    return new_image
```

```python
get_value_from_triple(im_1[10,10,:])
get_0_1_from_triple(im_1[10,10,:])
im_1_gray=convert_rgb_to_gray(im_1)
im_1_bw=convert_rgb_to_bw(im_1)
plt.subplot(1,3,1)
plt.imshow(im_1)  
plt.subplot(1,3,2)
plt.imshow(im_1_gray,cmap='gray')
plt.subplot(1,3,3)
plt.imshow(im_1_bw,cmap='gray')
plt.show()
plt.imsave('indir_gray.jpg',im_1_gray,cmap='gray')
plt.imsave('indir_bw.jpg',im_1_bw,cmap='gray')
```
